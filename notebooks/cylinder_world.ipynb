```python
import sys, os
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['figure.dpi'] = 150

sys.path.append(os.path.join('..'))
from cylinder_world import CylinderWorld
```

```python
world = CylinderWorld(resolution=256, n_coeffs=32, radius=2)
```

```python
fig, axs = plt.subplots(1, 3, sharey=True)
n_smpls = 3 * world.resolution + 1
xs = np.linspace(-0.5, 0.5, n_smpls)         # Move from x=-0.5 to 0.5
ys = np.linspace(-0.5, 0.5, n_smpls)         # Move from y=-0.5 to 0.5
thetas = np.linspace(-np.pi, np.pi, n_smpls) # Rotate 360°
Is, dIs = world.render_with_gradient(xs, ys, thetas)
ds = world.depth(xs, ys, thetas)

axs[0].set_title('Intensity')
axs[0].imshow(Is.T, extent=[0, 360, 0, n_smpls - 1])
axs[0].set_aspect('auto')

axs[1].set_title('Gradient')
axs[1].imshow(dIs.T, extent=[0, 360, 0, n_smpls - 1])
axs[1].set_aspect('auto')

axs[2].set_title('Depth')
axs[2].imshow(ds.T, extent=[0, 360, 0, n_smpls - 1])
axs[2].set_aspect('auto')

axs[0].set_xlabel('Angle $\\beta$')
axs[1].set_xlabel('Angle $\\beta$')
axs[2].set_xlabel('Angle $\\beta$')
axs[0].set_ylabel('Sample $i$');
```
